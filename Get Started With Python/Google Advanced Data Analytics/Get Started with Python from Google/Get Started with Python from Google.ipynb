```python
### Find all of the rows that have a null 'label' 
null_label_mask = df['label'].isnull()
null_label=df[null_label_mask]
# null_label.head()
# null_label.count()


### Find all of the rows that have a not null 'label' 
not_null_label_mask = df['label'].notnull()
not_null_label = df[not_null_label_mask]
# not_null_label.head()
# not_null_label.count()
```
